```python
data = data.remove_data(time_end=70,unit="picosecond")
```

```python
data.convert_property(what="time",unit="picosecond",family="time",inplace=True)
polarization = data.properties["totalpol"]
time = data.properties["time"]

fig, ax1 = plt.subplots(figsize=(12,4))

l = ["x","y","z"]
for i in range(3):
    ax1.plot(time,polarization[:,i],label=l[i],alpha=0.5)
ax1.plot(time,np.linalg.norm(polarization,axis=1)/np.sqrt(3),label="tot/sqrt(3)")


ax1.set_ylabel("polarization (a.u.)")

ax1.set_xlabel("time (ps)")

#labs = [l.get_label() for l in labels]
#ax1.legend(labels, labs, loc=0)
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
pos = data.to_ase()
V = pos[0].cell.volume
Qx = pos[0].cell[0,0] / V
Qy = pos[0].cell[1,1] / V
Qz = pos[0].cell[2,2] / V

nx = -4
ny = -4
nz = -4
Q = np.asarray([nx*Qx,ny*Qy,nz*Qz])
print(Q)
polarization = data.properties["totalpol"] + Q 

pos[0].get_chemical_symbols()
d = np.zeros((len(pos),3))
for i in range(len(pos)):
    p = pos[i]
    d[i] = (p.positions[0,:] - (p.positions[1,:] + p.positions[2,:])/2.)
    
#del pos
```

```python
data.convert_property(what="time",unit="picosecond",family="time",inplace=True)
time = data.properties["time"]

fig, ax = plt.subplots(1,3,figsize=(12,4))

ax[0].scatter(d[:,0],-polarization[:,0],label="x",alpha=0.4,s=0.5,c="red")
ax[1].scatter(d[:,1],-polarization[:,1],label="y",alpha=0.4,s=0.5,c="green")
ax[2].scatter(d[:,2],-polarization[:,2],label="z",alpha=0.4,s=0.5,c="blue")

#ax1.plot(time,d[:,0],label="dx",alpha=0.5)#,s=1)
#ax1.plot(time,polarization[:,1],label="y",alpha=0.5)#,s=1)
#ax1.plot(time,polarization[:,2],label="z",alpha=0.5)#,s=1)

#ax1.plot(time,np.linalg.norm(polarization,axis=1)/np.sqrt(3),label="tot/sqrt(3)")

for a in ax:
    a.set_xlabel("molecular orientation (a.u.)")
    a.set_ylabel("polarization (a.u.)")
    a.grid()
    a.legend()
    xlim = a.set_xlim()
    a.hlines(0,xlim[0],xlim[1],linestyle="--",color="black",alpha=0.7)
    ylim = a.set_ylim()
    a.vlines(0,ylim[0],ylim[1],linestyle="--",color="black",alpha=0.7)

#labs = [l.get_label() for l in labels]
#ax1.legend(labels, labs, loc=0)
#plt.legend()
plt.tight_layout()
#plt.grid()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12,4))

ax1.plot(data.properties["time"],data.properties["conserved"],label="energy",alpha=0.5)

ax1.set_ylabel("energy (a.u.)")

ax1.set_xlabel("time (ps)")

#labs = [l.get_label() for l in labels]
#ax1.legend(labels, labs, loc=0)
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
df = np.diff(data.velocities,axis=0)
dt = np.diff(data.properties["time"],axis=0)
dt = np.append(dt,dt[-1])

forces = np.zeros((df.shape[0]+1,df.shape[1]))
for i in range(df.shape[1]):
    dv = df[:,i]
    dv = np.append(dv,df[-1,i])
    forces[:,i] = dv/dt
forces = np.linalg.norm(forces,axis=1)
```

```python
dE = np.diff(data.properties["conserved"],axis=0)
dt = np.diff(data.properties["time"],axis=0)
dE = np.append(dE,dE[-1])
dt = np.append(dt,dt[-1])

driving = dE/dt
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))

ax[0].scatter(forces,driving,alpha=0.4,s=0.5,c="red")
ax[0].set_xlabel("forces (a.u.)")
ax[0].set_ylabel("driving (a.u.)")
ax[0].grid()

ax[1].scatter(np.linalg.norm(data.velocities,axis=1),driving,alpha=0.4,s=0.5,c="blue")
ax[1].set_xlabel("velocities (a.u.)")
ax[1].set_ylabel("driving (a.u.)")
ax[1].grid()

#plt.grid()
plt.tight_layout()
plt.show()
```
